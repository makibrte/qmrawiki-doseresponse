```python
import pandas as pd
import os
```

```python
data_dir = os.listdir('data/metadata/')
```

```python
for file_ in data_dir:
    if file_.split('.')[0].split('_')[-1] == 'metadata':
        df = pd.read_csv(f'data/metadata/{file_}')
        column_parts = df.columns[-2].split(' ')[2:]
        optim_params = df.columns[-2]
        value_parts = df[df.columns[-2]].values[0].replace('[', '').replace(']', '').split(',')

        print(f"Column Parts: {len(column_parts)}, Value Parts: {len(value_parts)}")

        for n, new_column in enumerate(column_parts):
            if n < len(value_parts):
                val = float(value_parts[n].strip())
                print(val)
                df[new_column] = val
            else:
                print(f"Index {n} out of range for value parts.")
        df.drop([optim_params], axis =1, inplace=True)
        #print(df)       
        df.to_csv(f'data/metadata/{file_}') 
        
        
```
